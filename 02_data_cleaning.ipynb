3. Analyzing Missing Values in Telecom Users dataframe

&ensp;&ensp;&ensp;&ensp;Before checking the missing values (NaN), it is necessary to replace all the blank spaces occurrences that this dataframe might have


```python
# Step 3: Analyzing Missing Values in columns (part 1)

# Replacing all blank spaces with NaN in Telecom dataframe
df_telecom_clean1 = df_telecom.replace(r'^\s*$', np.nan, regex=True)  
# ReGex that remove white spaces such as " " or "  " or "   " and so on. (Replace them by NaN)

# Finding missing values
print (df_telecom_clean1.isnull().sum()) # Output: One table showing the total of missing values (nan) for each Column in the given dataset
print ('''
As we can see above, the '''+'\033[1m'+'column TotalCharge'+'\033[0m'+''' has NaN values, which means that the value 
is not available there (missing).''')

# Checking the percentage of missing values.
TotalCh_Nan_per = (df_telecom_clean1['TotalCharges'].isnull().sum() / df_telecom_clean1.shape[0]*100).round(2) # Output: Percentage of missing values
TotalCh_Nan = df_telecom_clean1['TotalCharges'].isnull().sum() # Output: Count of missing values

print ('''
To be more precise, there are '''+'\033[1m'+'%s missing values'%(TotalCh_Nan)+'\033[0m'+''' in TotalCharge.
In other words, about '''+'\033[1m'+'%s'%(TotalCh_Nan_per)+'%'+'\033[0m'+' of the TotalCharge values are missing (%s/%s)' \
%(df_telecom_clean1['TotalCharges'].isnull().sum(), df_telecom_clean1.shape[0])+'''

As a result, the records (rows) with NaN values will be deleted from the dataframe since they do not represent more than 3%
of the whole data''')

# Code '\033[1m' and '\033[0m' allow to print a string in bold.

# .replace()    This function is used to replace a string, regex, list, dictionary, etc from a dataframe. 
#               In this case, " " or "  " or "   " will be replaced with NaN values/
# regex=True    This parameter indicates that the value that should be replaced is a regular expression.
# \s            This ReGex matches any whitespace character. In other words, it will find strings such as " ", "\r" or "\n"
# ^\s           This ReGex matches only whitespace characters that are found at the beginning.
# *             The asterisk (*) declares that "\s" is optional. (Could appear or not)
# \s$           This ReGex matches only whitespace characters that are found at the end.
# np.nan        This function is used to create a NaN value. 
```

5. Encoding all categorical variables

```python
# Step 5: Encoding categorical data (Part 1)

# Identifying categorical variables

# Get only the columns that are "Object" data-type from "Telecom Users" dataset
df_telecom_clean3_obj = df_telecom_clean3.dtypes[df_telecom_clean3.dtypes == np.object] 

print('''\nData-types of "Telecom Users" dataset:

%s   

As we can see below, the following columns are categorical and need to be encoded:

%s
#'''%(df_telecom_clean3.dtypes, df_telecom_clean3_obj))  
# Output 1: All data-types from "Telecom Users" dataset. 
# Output 2: Only object data-types from "Telecom Users" dataset. 

# .dtypes              This function is used to find the data-type in the dataframe.
#                      If [dtypes == np.object], it will find only the data-type that is object.
```

```python
# Step 5: Encoding categorical data (Part 2)

# Verifying uniques values for each categorical variable in the Telecom dataframe
np.set_printoptions(threshold=np.inf, precision=3, suppress=True, linewidth=200) # Print all the values of an array as a pretty table (all values in one row)
names_df_telecom_clean3_obj = list(df_telecom_clean3_obj.index)   # Get the names of the columns that are "object" data-type from "Telecom Users" dataset
df_telecom_clean4 = df_telecom_clean3.copy()                      # Create a copy of "Telecom Users" dataset

print ('Unique values for each categorical variable in the Original Telecom dataset:\n')
for item in names_df_telecom_clean3_obj:         # For-loop iterates over the values found in columns that are "Object" data-type in "Telecom Users" dataset
    print(item)
    print (df_telecom_clean3[item].unique())     # Print unique values for each column that is object in the given dataframe

# .unique()            This function is used to count distinct observations in the given column
# list(.index)         This function is used to get the names of the columns in a dataframe
# .copy()              This function is used to make a copy of one dataframe with indices and data
```

```python
# Step 5: Encoding categorical data (Part 3)

# Performing one-hot encoding to Binary features (Yes/No)
binary_colsYN = ['Partner','Dependents','PhoneService','PaperlessBilling'] # Binary features (Yes/No)

for i in binary_colsYN:                                  # For-loop iterates over the columns found in "binary_colsYN"
    df_telecom_clean4[i] = df_telecom_clean4[i].replace({"Yes":1,"No":0})   # Transforming categorical data into numbers

# Peforming one-hot encoding to other binary features
df_telecom_clean4["gender"]=df_telecom_clean4["gender"].replace({"Male":1,"Female":0}) # Transforming categorical data into numbers

print('\n"Telecom Users" dataset after performing one-hot encoding to binary-categorical features (All rows)(Truncated)')  
display(df_telecom_clean4) # Telecom dataframe after performing one hot encoding to binary features
```

```python
# Step 5: Encoding categorical data (Part 4)

# Performing one-hot encoding to the rest of the categorical variables

category_cols=['PaymentMethod','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection',
               'TechSupport','StreamingTV','StreamingMovies','Contract'] # Multi-categorical features

for col in category_cols:
    encoder = pd.get_dummies(df_telecom_clean4[col], drop_first=False)  # Transforming categorical data into numbers
    encoder = encoder.add_prefix("{}#".format(col))     # Add a delimitator "#" to each column created after performing one-hot encoder
    df_telecom_clean4.drop(col, axis=1, inplace=True)   # Delete the Multi-Categorical Columns
    df_telecom_clean4 = df_telecom_clean4.join(encoder) # Merge the Multi-categorical columns after performing one hot encoding into the Telecom dataframe

print('\n"Telecom Users" dataset after performing one-hot encoding to Multi-categorical features (All rows)(Truncated)') 
display(df_telecom_clean4) # Telecom dataframe after performing one hot encoding to multi-categorical features

# pd.get_dummies()    This function is used for data manipulation. 
#                     It converts categorical data into dummy or indicator variables (0 - 1)
# drop_first=False    This parameter is used to keep all columns after performing one-hot encoding
# inplace=True        This parameter is used to make changes in the original dataframe itself.
```

```python
# Creating a function to display countplots based on a set of features

chart1 = sns.color_palette("Reds", 20)      # Customize the color of the bars
chart1.reverse()                            # Start the graph with color Red

def displayCountPlot(list, df, rows=1, columns=3, figsize=(14,3.5)):
  
  """
    # Arguments:
      cat_list: array, List of features
      df: DataFrame, dataset
      rows: int, number of rows
      columns: int, number of columns
      figsize: figure size, e.g (14, 3.5)
  """

  fig, ax = plt.subplots(ncols=columns, figsize=figsize)  
  # This parameter is used to adjust the size of the graph and define the number of columns
  
  idx = 0                    # Index variable is initially 0.
  for col in list:           # For-loop iterates over all the existing columns in the dataframe given
    idx += 1                 # Output: idx = [idx + 1](increments 1 on each loop)   
    plt.subplot(rows, columns, idx)     # Print multiple countplots in accordance with the specified rows and columns
    ax = sns.countplot(x=df[col], data=df, palette=chart1)  # Create the countplot with the specified parameters.

    plt.xlabel('Churn', size=14)   # This X-Axis property allows us to customize the labels of X-axis
    plt.ylabel('')                 # This Y-Axis property allows us to customize the labels of Y-axis
    
    plt.xticks(size=13)     # This X-Axis property allows us to customize the values of the labels in X-axis
    plt.yticks(size=13)     # This Y-Axis property allows us to customize the values of the labels in Y-axis
     
    plt.subplots_adjust(hspace = 0.5)       # Change the amount of height reserved for white space between subplots
    ax.tick_params(labelleft=False)         # This axis parameter is used to hide the values of the labels in Y-Axis
    ax.set_title(col, fontsize=15)  # Add a title to the countplot

    print_rate(ax, df.shape[0]) # Call the function print_rate to place annotations.

  plt.tight_layout()   # Adjust the subplots to avoid misplacings or pairing.
  plt.show()           # Display the current figure that we are working on
    
# Creating a function to add annotations to the subplots   

def print_rate(ax, total):
  for bar in ax.patches:             # For-loop iterates over all the existing bars in the graph
    text = '{:.1f}% ({})'.format(100 * bar.get_height() / total, bar.get_height()) 
    # Action 1: Calculate the percentage of each bar, Action 2: fetch the total counts of each bar (feature)
    x = bar.get_x() + bar.get_width() / 2        # Get the value of X-Axis
    y = bar.get_y() + bar.get_height() * 0.5     # Get the value of Y-Axis
    ax.annotate(text, (x, y), size = 11.5, ha="center", va="center", color='white') 
    # Place the annotation with the specified parameters

# plt.subplots(rows, columns) This function is used to plot multiples charts in a single cell. 
#                             For example plt.sublots(1,2) will show "in one row two graphs"
# sns.countplot               This function is used to show the total counts in each categorical bin by using bars
# palette                     This parameter takes the palette name and represents the colors of the bars
# plt.subplots_adjust         This function is used to change the spacing between the subplots.
# plt.tight_layout()          This function is used to automatically adjust subplot parameters to give specified padding

# ax.patches    This syntax represents a bar and its corresponding background area.
# '{:.1f}%'     This syntax is used to format a number as float% with 1 digit of precision.
# get_height()  This parameter represents the value of the bar.
# ax.annotate   This method is used to create annotations.
# Size          This parameter is used to modify the font size value of the annotation
# ha            This parameter is used to center the annotation  (horizontally)
# va            This parameter is used to center the annotation  (vertically)
```
